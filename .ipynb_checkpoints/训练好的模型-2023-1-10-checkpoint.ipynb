```python
#导入需要的库
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
pd.options.display.max_columns = None #显示完整打印数据
pd.options.display.max_rows = None
plt.rcParams['font.sans-serif'] = ['Microsoft YaHei'] # 用来正常显示中文
plt.rcParams['axes.unicode_minus']=False    # 用来正常显示负号
plt.rcParams['figure.dpi'] = 150 # 修改图片分辨率
plt.rcParams.update({'font.size': 16}) # 修改图片整体字体大小
import numpy as np
import xgboost as xgb
from sklearn.tree import DecisionTreeClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.svm import SVC
from sklearn.impute import SimpleImputer
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
import lightgbm as lgb
from sklearn.model_selection import GridSearchCV 
import scikitplot as skplt
from sklearn.metrics import precision_score, recall_score, f1_score ,roc_curve, auc,confusion_matrix ,accuracy_score,roc_auc_score,auc,brier_score_loss
import warnings
warnings.filterwarnings("ignore") # 忽略警告
```

```python
# 读取数据集
data = pd.read_excel('肠癌18月无进展RF插补.xlsx')
# 查看数据前5行
data.head()
```

```python
data_copy = data.copy()
```

```python
data_copy.isnull().sum()
```

数据集划分

```python
# 分离特征和标签
X = data_copy.drop(['Tumor progression in18 months'],axis=1)
Y = data_copy['Tumor progression in18 months']
```

```python
# 数据分割
# 8比2划分训练集，测试集，设置随机种子random_state，保证实验能够复现
x_train, x_test, y_train, y_test = train_test_split(X,
                                                    Y,
                                                    test_size=0.2,
                                                    stratify=Y,
                                                    random_state=1877)
```

```python
# 训练测试特征形状
x_train.shape,x_test.shape
```

```python
x_train.head()
```

```python
x_test.head()
```

最佳算法确认

```python
# 定义五种评价指标
def try_different_method(model):
    model.fit(x_train,y_train)
    yuce = model.predict(x_test)
    predict_proba = model.predict_proba(x_test)[:, 1]
    print('测试集')
    precision = precision_score(y_test, yuce)
    recall = recall_score(y_test, yuce)
    f1score = f1_score(y_test, yuce)
    accuracy=accuracy_score(y_test, yuce)
    y1_test_score = model.predict_proba(x_test)
    fpr, tpr, thresholds = roc_curve(y_test, y1_test_score[:, 1])
    AUC = auc(fpr, tpr)
    brier_loss = brier_score_loss(y_test,predict_proba)
    
    print("AUC:      ", '%.4f'%float(AUC))
    print("Accuracy: ", '%.4f'%float(accuracy))
    print("Precision:", '%.4f'%float(precision))
    print("Recall:   ",'%.4f'%float(recall))
    print("f1_score：", '%.4f'%float(f1score))
#   print("brier_score_loss：", '%.4f'%float(brier_loss)) 
```

```python
# GSXGB
GSXGB = xgb.XGBClassifier(verbosity=0,
                            n_jobs=-1,
                            random_state=42,
                            max_depth=3,
                            colsample_bytree=0.5,
                            n_estimators=1100,
                            learning_rate=0.001)
print('GSXGB模型评分如下：')
try_different_method(GSXGB)
```

```python
# GSXGB模型混淆矩阵
plt.rcParams['figure.dpi'] = 150 # 修改图片分辨率
plt.rcParams.update({'font.size': 16})
y_pred8 = GSXGB.predict(x_test)
matrix = confusion_matrix(y_test, y_pred8)
dataframe = pd.DataFrame(matrix, index=['0','1'], columns=['0','1'])
sns.heatmap(dataframe, annot=True, cbar=None, cmap="Oranges_r",fmt='.5g',square=True)
plt.title("GSXGB"), plt.tight_layout()
plt.ylabel("True Class"), plt.xlabel("Predicted Class")
plt.savefig('GSXGB.png',dpi=300,bbox_inches = 'tight')
plt.show()
```
