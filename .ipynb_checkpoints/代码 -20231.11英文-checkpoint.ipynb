```python
# 随机森林模型填补缺失值
# 先填补分类型变量
data_copy = data.copy() # 复制一份数据
colume = ['Age', 'Gender', 'Nationalities', 'Height', 'Weight', 'BMI', 'Number of previous chemotherapy cycles', 'Tumor surgery', 
          'Primary tumor site', 'Tumor location', 'Histological type', 'Maximum tumor diameter(cm)', 'NLM', 
           'N', 'M', 'Tumor stage', 'Targeted therapy', 'Number of chemotherapy',
          'Completion of chemotherapy', 'Radiation therapy', 'FH of GI cancer', 'Hypertension', 'Diabetes', 'CHD', 'BPH', 'Nephrolithiasis', 
          'Gallstone', 'Chronic gastritis ', 'Duodenal ulcer', 'Cholecystitis', 'Fatty liver', 'Cirrhosis', 'Hepatic cyst', 'Renal cyst', 
          'Hyperlipidemia', 'Gout', 'Cerebral peduncle', 'Nausea and vomiting ', 'Dysphagia', 'Myelosuppression', 'Abnormal liver function',
          'NRS score', 'ECOG score', 'CEA', 'AFP', 'CA125', 'CA153', 'CA199', 'CA242', 'CA50', 'CA724', 'FER', 'TK1', 'Hsp90α',
          'WBC', 'hsCRP', 'MONO%', 'MONO', 'RDW-SD', 'MCV', 'MCH', 'MCHC', 'RBC', 'HCT', 'LYMPH%', 'LYMPH', 'BASO%', 'BASO', 'EOS%', 'EOS',
          'HGB', 'PLT', 'NEUT%', 'NEUT', '5NT', 'AFU', 'αHBDH', 'GGT', 'ALB', 'A/G', 'ALT', 'CHE', 'LDL-c', 'CO2', 'Ca', 'TG', 'HDL-c', 
          'eGFR', 'AST/ALT', 'Cys_C', 'Cr', 'CK', 'K', 'IBIL', 'ALP', 'P', 'Cl', 'Mg', 'Na', 'Urea', 'U/C', 'UA', 'GLU', 'PA', 'GLB', 'LDH',
          'OSM', 'AST', 'HCY', 'ADA', 'AG', 'ApoA', 'ApoB', 'Lpa', 'DBIL', 'CHOL', 'TBIL', 'TBA', 'TP', 'Tumor progression in18 months']
data_copy[colume] = data_copy[colume].replace(np.nan,0)
sindex = np.argsort(data_copy.isna().sum().values.tolist()) #将有缺失值的列按缺失值的多少由小到大
# 进入for循环进行空值填补
for i in sindex:                                          # 按空值数量,从小到大进行排序来遍历
    if data_copy.iloc[:,i].isna().sum() == 0:             # 将没有空值的行过滤掉
        continue                                          # 直接跳过当前的for循环
    df = data_copy                                        # 复制df数据
    fillc = df.iloc[:,i]                                  # 将第i列的取出，之后作为y变量
    df = df.iloc[:,df.columns != df.columns[i]]           # 除了有这列以外的数据，之后作为X
    df_0 = SimpleImputer(missing_values=np.nan,           # 将df的数据全部用0填充
                         strategy="constant",
                         fill_value=0).fit_transform(df)
    Ytrain = fillc[fillc.notnull()]                       # 在fillc列中,不为NAN的作为Y_train
    Ytest = fillc[fillc.isnull()]                         # 在fillc列中,为NAN的作为Y_test 
    Xtrain = df_0[Ytrain.index,:]                         # 在df_0中(已经填充了0),中那些fillc列不为NAN的行作为Xtrain
    Xtest = df_0[Ytest.index,:]                           # 在df_0中(已经填充了0),中那些fillc等于NAN的行作为X_test
    
    rfc = RandomForestClassifier(random_state=2022)
    rfc.fit(Xtrain,Ytrain)
    Ypredict = rfc.predict(Xtest)                         #Ytest为了定Xtest,以最后预测出Ypredict
    
    data_copy.loc[data_copy.iloc[:,i].isnull(),data_copy.columns[i]] = Ypredict    
    # 将data_copy中data_copy在第i列为空值的行,第i列,改成Ypredict
```

```python
# 再填补连续型变量
data_copy[colume] = data[colume]
sindex = np.argsort(data_copy.isna().sum().values.tolist()) #将有缺失值的列按缺失值的多少由小到大
# 进入for循环进行空值填补
for i in sindex:                                          # 按空值数量,从小到大进行排序来遍历
    if data_copy.iloc[:,i].isna().sum() == 0:             # 将没有空值的行过滤掉
        continue                                          # 直接跳过当前的for循环
    df = data_copy                                        # 复制df数据
    fillc = df.iloc[:,i]                                  # 将第i列的取出，之后作为y变量
    df = df.iloc[:,df.columns != df.columns[i]]           # 除了有这列以外的数据，之后作为X
    df_0 = SimpleImputer(missing_values=np.nan,           # 将df的数据全部用0填充
                         strategy="constant",
                         fill_value=0).fit_transform(df)
    Ytrain = fillc[fillc.notnull()]                       # 在fillc列中,不为NAN的作为Y_train
    Ytest = fillc[fillc.isnull()]                         # 在fillc列中,为NAN的作为Y_test 
    Xtrain = df_0[Ytrain.index,:]                         # 在df_0中(已经填充了0),中那些fillc列不为NAN的行作为Xtrain
    Xtest = df_0[Ytest.index,:]                           # 在df_0中(已经填充了0),中那些fillc等于NAN的行作为X_test
    
    rfc = RandomForestRegressor(random_state=2022)
    rfc.fit(Xtrain,Ytrain)
    Ypredict = rfc.predict(Xtest)                         #Ytest为了定Xtest,以最后预测出Ypredict
    
    data_copy.loc[data_copy.iloc[:,i].isnull(),data_copy.columns[i]] = Ypredict    
    # 将data_copy中data_copy在第i列为空值的行,第i列,改成Ypredict
```

# 4.七种机器学习算法对比

```python
# 定义五种评价指标
def try_different_method(model):
    model.fit(x_train,y_train)
    yuce = model.predict(x_test)
    predict_proba = model.predict_proba(x_test)[:, 1]
    print('测试集')
    precision = precision_score(y_test, yuce)
    recall = recall_score(y_test, yuce)
    f1score = f1_score(y_test, yuce)
    accuracy=accuracy_score(y_test, yuce)
    y1_test_score = model.predict_proba(x_test)
    fpr, tpr, thresholds = roc_curve(y_test, y1_test_score[:, 1])
    AUC = auc(fpr, tpr)
    brier_loss = brier_score_loss(y_test,predict_proba)
    
    print("AUC:      ", '%.4f'%float(AUC))
    print("Accuracy: ", '%.4f'%float(accuracy))
    print("Precision:", '%.4f'%float(precision))
    print("Recall:   ",'%.4f'%float(recall))
    print("f1_score：", '%.4f'%float(f1score))
#   print("brier_score_loss：", '%.4f'%float(brier_loss)) 
```

```python
# 建立不同的模型
# XGBoost
model_XGB = xgb.XGBClassifier(verbosity=0,
                            n_jobs=-1,
                            random_state=42,
                            max_depth=10,
                            colsample_bytree=0.6,
                           n_estimators=1000,
                            learning_rate=0.001)
# 决策树
model_DT = DecisionTreeClassifier(random_state=0,
                                  max_depth=6)
# 梯度提升树（GBDT）
model_GBDT = GradientBoostingClassifier(random_state=0,
                                        max_depth=10,
                                        n_estimators=500,
                                       learning_rate=0.1)
# 贝叶斯
model_NB = GaussianNB()
# 极端森林
model_ET =  ExtraTreesClassifier(random_state=0,
                                 max_depth=4,
                                 n_estimators=10,
                                 #class_weight={0:1,1:1.35}
                                 )
# AdaBoost
model_AB = AdaBoostClassifier(random_state=0,
                              n_estimators=400,
                             #learning_rate=0.95
                              )
# SVM
model_SVM = SVC(probability=True,
               random_state=0,
                kernel='linear',
                max_iter=90000
               )
```

```python
# 训练集五折交叉验证,AUC
from sklearn.model_selection import cross_val_score,StratifiedKFold
stratifiedkf=StratifiedKFold(n_splits=5)
models = []
models.append(model_XGB)
models.append(model_DT)
models.append(model_GBDT)
models.append(model_NB)
models.append(model_ET)
models.append(model_AB)
models.append(model_SVM)
cv_auc = []
for model in models :
    cv_auc.append(cross_val_score(model, x_train, y_train, scoring = "roc_auc", cv = stratifiedkf, n_jobs=-1))
cv_all_auc = pd.DataFrame(columns=["XGB","DT","GBDT","NB","ET","AB","SVM"])
c=0
for cv_result in cv_auc:
    cv_all_auc.iloc[:,c]=cv_result
    c=c+1
# 五折交叉平均auc
cv_means = []
for cv_result in cv_auc:
    cv_means.append(cv_result.mean())
auc_means = pd.DataFrame({"AUC":cv_means,"Model":["XGB","DT","GBDT","NB","ET","AB","SVM"]})
auc_means
```

```python
# 五折交叉验证可视化
plt.figure(figsize=(6,4))
y1 = auc_means['AUC']
x1 = auc_means['Model']
plt.bar(x1,y1,width=0.7,color = ['olivedrab','y','salmon','skyblue','lightpink','c','r'])
plt.xlabel('模型',fontsize=15) # x 轴
plt.ylabel('AUC',fontsize=15) # y轴
plt.title('五折交叉验证结果比较',fontsize=20) # 标题
plt.yticks(np.arange(0, 1.2, step=0.2))
#为每个条形图添加数值标签
for x,y in enumerate(y1):
    plt.text(x,y+0.02,'%.4f'%float(y),ha='center',fontsize=12)
plt.savefig('五折交叉验证结果比较.png',dpi=300,bbox_inches = 'tight')
#显示图形
plt.show()
```

```python
# 输出测试集评价指标
print('XGBoost模型评分如下：')
try_different_method(model_XGB)
print('决策树模型评分如下：')
try_different_method(model_DT)
print('梯度提升树模型评分如下：')
try_different_method(model_GBDT)
print('贝叶斯模型评分如下：')
try_different_method(model_NB)
print('极端森林模型评分如下：')
try_different_method(model_ET)
print('AdaBoost模型评分如下：')
try_different_method(model_AB)
print('SVM模型评分如下：')
try_different_method(model_SVM)
```

```python
# ROC曲线
def roc_curve_and_score(y_train, pred_proba):
    fpr, tpr, _ = roc_curve(y_train.ravel(), pred_proba.ravel())
    roc_auc = roc_auc_score(y_train.ravel(), pred_proba.ravel())
    return fpr, tpr, roc_auc
plt.figure(figsize=(6, 4))
plt.grid()
fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_XGB.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='r', lw=2,
         label='AUC XGBoost={0:.4f}'.format(roc_auc))

fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_DT.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='b', lw=2,
         label='AUC DecisionTree={0:.4f}'.format(roc_auc))

fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_GBDT.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='y', lw=2,
         label='AUC GradientBoosting={0:.4f}'.format(roc_auc))

fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_NB.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='lime', lw=2,
         label='AUC GaussianNB={0:.4f}'.format(roc_auc))


fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_ET.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='c', lw=2,
         label='AUC ExtraTrees={0:.4f}'.format(roc_auc))



fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_AB.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='m', lw=2,
         label='AUC AdaBoost={0:.4f}'.format(roc_auc))

fpr, tpr, roc_auc = roc_curve_and_score(y_test, model_SVM.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='green', lw=2,
         label='AUC SVM={0:.4f}'.format(roc_auc))

fpr, tpr, roc_auc = roc_curve_and_score(y_test, GSXGB.predict_proba(x_test)[:, 1])
plt.plot(fpr, tpr, color='slategrey', lw=2,
         label='AUC GSXGB={0:.4f}'.format(roc_auc))

plt.plot([0, 1], [0, 1],lw=1, linestyle='--')
plt.legend(loc="lower right",fontsize=8.7)
plt.title('ROC曲线')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.0])
plt.xlabel('假阳性率')
plt.ylabel('真阳性率')
plt.savefig('ROC曲线.png',dpi=300,bbox_inches = 'tight')
plt.show()
```
